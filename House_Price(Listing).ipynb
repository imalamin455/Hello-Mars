The list house Prices = [140, 550, 480] contains the prices of three houses in thousands of dollars. • Update the price of the second element in house Prices to 175

```python
house_Prices = [140, 550, 480]
# update second element
house_Prices[1] = 175
Print = house_Prices
print(Print)
```

•	Add a new element to the end of house Prices with a value of 260

```python
house_Prices = [140, 550, 480]
# add new element
house_Prices.append(260)
Print = house_Prices
print(Print)
```

•	Remove the first element from housePrices

```python
house_Prices = [140, 550, 480, 260] # Re-initialize the list from the previous state for clarity.
# Remove the first element
house_Prices.pop(0)
Print = house_Prices
print(Print)
```

•	Remove 480 from housePrices

```python
house_Prices = [550, 480, 260] # Re-initialize the list from the previous state for clarity.
# Remove 480 from housePrices
house_Prices.remove(480)
Print = house_Prices
print(Print)
```

•	You can estimate the average of the list housePrices

```python
# Calculate the average of house_Prices
# Using the last state of house_Prices after all modifications: [550, 260]
house_Prices_for_avg =  [140, 550, 480, 260]
average_price = sum(house_Prices_for_avg) / len(house_Prices_for_avg)
print(f"The average house price is: {average_price}")
```

•	Sort the list from lower to higher order, or higher to lower order

```python
house_Prices = [140, 550, 480, 260]

# Sort from lower to higher order (ascending)
house_Prices_asc = list(house_Prices) # Create a copy to sort
house_Prices_asc.sort()
print(f"Sorted (ascending): {house_Prices_asc}")

# Sort from higher to lower order (descending)
house_Prices_desc = list(house_Prices) # Create another copy
house_Prices_desc.sort(reverse=True)
print(f"Sorted (descending): {house_Prices_desc}")
```

•	Check a condition: the highest house price is above 300k

```python
# Assuming house_Prices contains the current list of prices.
house_Prices = [140, 550, 480, 260]

highest_priceee = max(house_Prices)

if highest_priceee > 300:
    print(f"The highest house price ({highest_priceee}k) is above 300k.")
else:
    print(f"The highest house price ({highest_priceee}k) is not above 300k.")
```

Create a dictionary with two list items, where car_brand list shows 4 car brands with respective temperatures in the temperatures list.

```python
car_brands = ["Toyota", "Honda", "BMW", "Tesla"]
temperatures = [20, 25, 30, 35]

data_dict = {
    "Car Brands": car_brands,
    "Temperatures": temperatures,
}

print(data_dict)
```

•	Add ‘age’ to the following dictionary, where age contains driver ages: 18, 25, 40, 60



```python
data_dict["age"] = [18, 25, 40, 60]
print(data_dict)
```

•Removes the list with key Temperatures

```python
data_dict.pop("Temperatures", None)
print(data_dict)
```

•	Change the first element of the list “age” from 18 to 19:

```python
data_dict["age"][0] = 19
print(data_dict)
```

•	Get the average value of the list age from the dictionary:

```python
average_age = sum(data_dict["age"])/len(data_dict["age"])
print(f"The average age is: {average_age}")
```

•	Get the car brand names where temperatures are equal to or below 25:

```python
car_brands_below_25 = [
    data_dict['Car Brands'][i]
    for i in range(len(data_dict['Temperatures']))
    if data_dict['Temperatures'][i] <= 25
]
print(f"Car brands with temperatures equal to or below 25: {car_brands_below_25}")
```
