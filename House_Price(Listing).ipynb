•	Sort the list from lower to higher order, or higher to lower order

```python
house_Prices = [140, 550, 480, 260]

# Sort from lower to higher order (ascending)
house_Prices_asc = list(house_Prices) # Create a copy to sort
house_Prices_asc.sort()
print(f"Sorted (ascending): {house_Prices_asc}")

# Sort from higher to lower order (descending)
house_Prices_desc = list(house_Prices) # Create another copy
house_Prices_desc.sort(reverse=True)
print(f"Sorted (descending): {house_Prices_desc}")
```

•	Check a condition: the highest house price is above 300k

```python
# Assuming house_Prices contains the current list of prices.
house_Prices = [140, 550, 480, 260]

highest_priceee = max(house_Prices)

if highest_priceee > 300:
    print(f"The highest house price ({highest_priceee}k) is above 300k.")
else:
    print(f"The highest house price ({highest_priceee}k) is not above 300k.")
```

Create a dictionary with two list items, where car_brand list shows 4 car brands with respective temperatures in the temperatures list.

```python
car_brands = ["Toyota", "Honda", "BMW", "Tesla"]
temperatures = [20, 25, 30, 35]

data_dict = {
    "Car Brands": car_brands,
    "Temperatures": temperatures,
}

print(data_dict)
```

•	Add ‘age’ to the following dictionary, where age contains driver ages: 18, 25, 40, 60



```python
data_dict["age"] = [18, 25, 40, 60]
print(data_dict)
```

```python
average_age = sum(data_dict["age"])/len(data_dict["age"])
print(f"The average age is: {average_age}")
```

•	Get the car brand names where temperatures are equal to or below 25:

```python
car_brands_below_25 = [
    data_dict['Car Brands'][i]
    for i in range(len(data_dict['Temperatures']))
    if data_dict['Temperatures'][i] <= 25
]
print(f"Car brands with temperatures equal to or below 25: {car_brands_below_25}")
```

Practice Problem:
You are working as a data analyst for a retail company that collects various types of customer and product data. Your task is to classify the data, store it in appropriate Python data structures, and perform some queries.
Data Collected:
•	Customer Names: "Alice", "Bob", "Charlie", "David"
•	Customer Satisfaction Ratings (1-5): 3, 5, 2, 4
•	Product Prices (in USD): 19.99, 45.50, 30.00, 89.99
•	Purchase Dates: "2023-05-21", "2023-06-10", "2023-07-15", "2023-08-01"
•	Product Categories: "Electronics", "Clothing", "Furniture", "Toys"
•	Customer Ages: 25, 34, 40, 19
•	Order Quantities: 2, 1, 5, 3
•	Product Ratings (out of 5): 4.5, 3.8, 4.0, 4.9
Task 1: Create Python lists to store each category of data.

```python
customer_names = ["Alice", "Bob", "Charlie", "David"]
customer_satisfaction_ratings = [3, 5, 2, 4]
product_prices = [19.99, 45.50, 30.00, 89.99]
purchase_dates = ["2023-05-21", "2023-06-10", "2023-07-15", "2023-08-01"]
product_categories = ["Electronics", "Clothing", "Furniture", "Toys"]
customer_ages = [25, 34, 40, 19]
order_quantities = [2, 1, 5, 3]
product_ratings = [4.5, 3.8, 4.0, 4.9]
```

Task 2: Construct a nested dictionary to store the data, with keys representing the category names and values storing respective lists. Name the dictionary “data_dict”

```python
data_dict = {
    "Customer Names": customer_names,
    "Customer Satisfaction Ratings": customer_satisfaction_ratings,
    "Product Prices": product_prices,
    "Purchase Dates": purchase_dates,
    "Product Categories": product_categories,
    "Customer Ages": customer_ages,
    "Order Quantities": order_quantities,
    "Product Ratings": product_ratings
}

print(data_dict)
```

Task 3: Querying the Data:
•	Find and print the highest and lowest product prices.

```python
product_ratings = data_dict['Product Ratings']
#Check if any product has a pefect rating (5.0)
Perfect_Rating = 5.0 in product_ratings

#Get the highest rating given by the customers
highest_rating = max(product_ratings)

print(f" Are there any products with a perfect rating (5.0)? {Perfect_Rating}")
print(f"The highest product rating is: {highest_rating}")
```

•	Calculate the total revenue by multiplying product prices with order quantities.

```python
product_prices = data_dict['Product Prices']
order_quantities = data_dict['Order Quantities']

total_revenue = sum(price * quantity for price, quantity in zip(product_prices, order_quantities))
print(f"The total revenue is: ${total_revenue:.2f}")
```

•	Retrieve all customer names who have a satisfaction rating of 4 or above.
